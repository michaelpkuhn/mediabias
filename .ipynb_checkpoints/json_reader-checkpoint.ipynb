```python
import datetime
import pandas as pd

json_data = '../mediabias_main/congresstweets/data/'
day = '2017-06-21'
day_dt = datetime.datetime.strptime(day, '%Y-%m-%d')
break_time = datetime.datetime(2018, 6, 22)

frames = []

while day_dt < break_time:
    try:
        day_str = day_dt.strftime('%Y-%m-%d')
        day_ref = json_data+day_str+'.json'
        day_df = pd.read_json(day_ref)
        frames.append(day_df)
        day_dt += datetime.timedelta(days=1)
    except Exception as e:
        print(e)
        day_dt += datetime.timedelta(days=1)
        pass

len(frames)

result_df = pd.concat(frames)

link_df = result_df[result_df['text'].str.contains('www.')]

external_df = link_df.copy(deep=True).set_index('id')

def link_filter(df, url):
    df = df[~df.text.str.contains(url)]
    return df

url_filter = ['twitter.com', 'pscp.tv']

for link in url_filter:
    external_df = link_filter(external_df, link)

pd.set_option('display.max_colwidth', -1)

regex = '(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?'
url_extract = external_df.text.str.extract(regex)
url_extract.drop(columns=0, inplace=True)
len(url_extract)

final_df = external_df.join(url_extract)
```
